 # Question: 
 what is the distance to all these galaxies assuming that their apparent brightness (m) is equal to 16?

```
#Distance Modulus: m - M = 5 - 5log(d) 
#Form we want: d = 10^((m-M-5)/-5)
m = [-10,-12,-14,-16,-18,-20,-22]

def DistanceModulus(m,M):
  d = 10**((m-M+5)/5)
  return d
for element in m:
  print(DistanceModulus(16,element))
```
